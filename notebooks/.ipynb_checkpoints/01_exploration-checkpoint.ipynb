```python
import pandas as pd
import seaborn as sns

df = sns.load_dataset("tips")

df.head()
```

```python
df.info()
df.describe()
df["smoker"].value_counts()
```

```python
df.groupby("smoker")["total_bill"].mean()
```

```python
import matplotlib.pyplot as plt

df.groupby("smoker", observed= True)["total_bill"].mean().plot(kind="bar")
plt.title("Average cost by Smoking Status")
plt.show()
```

```python
df.groupby (["sex", "smoker"])["total_bill"].mean()
```
