```python
#Start day 4
import pandas as pd
import numpy as np

df = pd.read_csv('../data/diabetic_data.csv')
df = df.replace('?', np.nan)

df['readmitted_flag'] = df['readmitted'].apply(
    lambda x: 1 if x == '<30' else 0
)
```

```python
df['long_stay'] = df['time_in_hospital'].apply(
    lambda x:1 if int(x) >= 7 else 0
)
```

```python
#checking to see data 
df.groupby('long_stay')['readmitted_flag'].mean()
```

```python
df ['num_medications'] = pd.to_numeric(df['num_medications'], errors='coerce')
df['high_med_burden'] = df['num_medications'].apply(
    lambda x: 1 if x >=15 else 0
)
```

```python
df.groupby('high_med_burden')['readmitted_flag'].mean()
```

```python
df['number_outpatient'] = pd.to_numeric(df['number_outpatient'], errors='coerce')
df['number_inpatient'] = pd.to_numeric(df['number_inpatient'], errors='coerce')
df['number_emergency'] = pd.to_numeric(df['number_emergency'], errors='coerce')

df['high_utilizer'] = (
    df['number_outpatient'] + df['number_inpatient'] + df['number_emergency']
).apply(lambda x: 1 if x >= 3 else 0)
```

```python
df.groupby('high_utilizer')['readmitted_flag'].mean()
```

```python
df['risk_score'] = (
    df['long_stay'] +
    df['high_med_burden'] +
    df['high_utilizer']
)
```

```python
df.groupby('risk_score')['readmitted_flag'].mean()
```

```python
import matplotlib.pyplot as plt

df.groupby('risk_score')['readmitted_flag'].mean().plot(kind='bar')
plt.title ("Readmission Risk by Risk Score")
plt.ylabel("Readmission Rate")
plt.show()
```

```python
df['age_risk'] = df['age'].apply(
    lambda x: 1 if x in ['[20-30)', '[70-80)', '[80-90)', '[90-100)'] else 0
)

df['risk_score_v2'] = df['risk_score'] + df['age_risk']
```

```python
df.groupby('risk_score_v2')['readmitted_flag'].mean()
```

Mash up Risk score with combination of legnth of stay, medication burden, and 
prior utilization demonstrated a stepwise increase in 30-day readmission risk. 
patient with the highest risk score had significantly elevated readmission rates, suggesting that simple rule-based stratification maybe effective for early identification of high-risk patients
